```python
import requests
import json

def fetch_block_children(block_id, headers):
    children = []
    has_more = True
    start_cursor = None

    while has_more:
        url = f"https://api.notion.com/v1/blocks/{block_id}/children"

        if start_cursor:
            url += f"?start_cursor={start_cursor}"

        response = requests.get(url, headers=headers)
        response_json = response.json()

        children.extend(response_json["results"])
        has_more = response_json["has_more"]
        start_cursor = response_json["next_cursor"]

    # Fetch the children of each child block
    for child in children:
        if child["has_children"]:
            child["children"] = fetch_block_children(child["id"], headers)

    return children


def fetch_page_content(page_id, headers):
    # Fetch the top-level blocks in the page
    blocks = fetch_block_children(page_id, headers)

    # Recursively fetch the children of each block
    for block in blocks:
        if block["has_children"]:
            block["children"] = fetch_block_children(block["id"], headers)

    return blocks

headers = {
    "Authorization": f"Bearer {NOTION_API_KEY}",
    "Content-Type": "application/json",
    "Notion-Version": "2022-06-28"
}

page_id = "5f1cf603-e5a6-4b22-8e72-870c517738bc"
page_content = fetch_page_content(page_id, headers)

# Now page_content contains all the content from the page
```

```python
tickers = ["MSFT", "AAPL", "AVGO", "TXN", "DOCN"]

def calculate_latest_price():
    return "120 USD" 

def calculate_change():
    return "-4 %"

def create_content(ticker, json_file):
    """
    Creates a content dictionary for a given ticker.
    
    Parameters:
    ticker (str): The ticker symbol.
    json_file (dict): The JSON file containing ticker data.

    Returns:
    dict: A dictionary containing sentiment counts and summary.
    """
    if ticker not in json_file:
        return f'Ticker {ticker} not found in json file.'

    ticker_data = json_file[ticker]
    ticker_data["summaries"] = [i for i in ticker_data["summaries"] if i != "All photographs subject to copyright."]
    summary = " ".join(ticker_data["summaries"])
    good = ticker_data["sentiments"].count("positive")
    neutral = ticker_data["sentiments"].count("neutral")
    bad = ticker_data["sentiments"].count("negative")

    # Calculate latest price and change dynamically
    latest_price = calculate_latest_price()
    change = calculate_change()

    return {
        "G": f"G: {good}",
        "N": f"N: {neutral}",
        "B": f"B: {bad}",
        "Summary": summary,
        "Latest Price": latest_price,
        "Change": change,
    }
```
